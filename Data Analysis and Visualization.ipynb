```python
t = glob('./data/train/{}/*.jpg'.format('PNEUMONIA'))
t1 = glob('./data/train/{}/*.jpg'.format('NORMAL'))
t2 = glob('./data/train/{}/*.jpg'.format('COVID19'))
```

```python
#placing title as the label (of the target or dependent column in our dataset)
#placing xlabel as the shape of that image inorder to get an idea of it

fig, ax = plt.subplots(3, 3)
fig.set_size_inches(10,10)
ax[0,0].imshow(plt.imread(t[0]))
ax[0,0].set_title("PNEUMONIA")
ax[0,0].set_xlabel(("Shape is:",plt.imread(t[0]).shape))
ax[0,1].imshow(plt.imread(t1[0]))
ax[0,1].set_title("NORMAL")
ax[0,1].set_xlabel(("Shape is:",plt.imread(t1[0]).shape))
ax[0,2].imshow(plt.imread(t2[0]))
ax[0,2].set_title("COVID19")
ax[0,2].set_xlabel(("Shape is:",plt.imread(t2[0]).shape))

ax[1,0].imshow(plt.imread(t[1]))
ax[1,0].set_title("PNEUMONIA")
ax[1,0].set_xlabel(("Shape is:",plt.imread(t[1]).shape))
ax[1,1].imshow(plt.imread(t1[1]))
ax[1,1].set_title("NORMAL")
ax[1,1].set_xlabel(("Shape is:",plt.imread(t1[1]).shape))
ax[1,2].imshow(plt.imread(t2[1]))
ax[1,2].set_title("COVID19")
ax[1,2].set_xlabel(("Shape is:",plt.imread(t2[1]).shape))

ax[2,0].imshow(plt.imread(t[2]))
ax[2,0].set_title("PNEUMONIA")
ax[2,0].set_xlabel(("Shape is:",plt.imread(t[2]).shape))
ax[2,1].imshow(plt.imread(t1[2]))
ax[2,1].set_title("NORMAL")
ax[2,1].set_xlabel(("Shape is:",plt.imread(t1[2]).shape))
ax[2,2].imshow(plt.imread(t2[2]))
ax[2,2].set_title("COVID19")
ax[2,2].set_xlabel(("Shape is:",plt.imread(t2[2]).shape))
plt.show()
```
